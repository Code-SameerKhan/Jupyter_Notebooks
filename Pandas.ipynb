```python
import pandas as pd
def adding(new_row,row_index,data):
    data1 = data.head(1)
    data.concat([data1,new_row],sort = True).reset_index(drop = True)
    
student_data = {
    'phone_number' : [999999121,999999122,999999123,999999124,999999125,999999126],
    'course_name' : ['python','data_science','digital_marketing','big_data','AI','python'],
    'enrolled' : ['yes','yes','no','no','yes','yes']
}
data = pd.DataFrame(student_data)
new_row = {
    'phone_number':[999999127,999999128],
    'course_name':['C++','java'],
    'enrolled':['yes','no']
}
row_index = 2
adding(new_row,row_index,data)
```

```python
#function to add a new row in b/n an existing dataframe at a particular index
def insert_row(row_number,df,row_value):
    start_upper = 0
    end_upper = row_number
    start_lower = row_number
    end_lower = df.shape[0] #extracting number of rows in a dataframe
    upper_half = [i for i in range(start_upper,end_upper)]
    lower_half = [i for i in range(start_lower,end_lower)]
    lower_half = [x+1 for x in lower_half] #[2,3,4]-->[3,4,5]
    index_ = upper_half+lower_half
    df.index = index_
    df.loc[row_number] = row_value
    df = df.sort_index()
    return df
```

```python
import pandas as pd
df = pd.DataFrame({
    'Key':['K0','K1','K3','K4','K5'],
    'A' : ['A0','A1','A3','A4','A5']
})
df
```

```python
df2 = pd.DataFrame({
    'Key': ['K0','K1','K2'],
    'B': ['B0','B1','B2']
})
df2
```

```python
df.join(df2,lsuffix='_first',rsuffix='other') #by defaut it is 'left join'->left table + common values
```

```python
df.join(df2,lsuffix='_first',rsuffix='other',how = 'inner') #inner -> joins only commmon values of the tables
                                                            # suffix is added to avoid key overlapping
```

```python
df.join(df2,lsuffix='_first',rsuffix='other',how = 'right') #right -> common values + common values
```

```python
df.join(df2,lsuffix='_first',rsuffix='other',how = 'outer') # outer gives all values of both tables
```

```python
d1 = df.set_index('Key') #joining tables on the basis of keys by first making them indexes
d2 = df2.set_index('Key')
d1.join(d2)
```

```python
df.join(df2.set_index('Key'), on = 'Key') # 'on' in 'df' gives program to arrange accoring to which column in this case-> 'Key'
```
