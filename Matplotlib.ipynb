```python
import matplotlib.pyplot as plt
%matplotlib inline
days = [1,2,3,4,5,6,7]
max_t = [50,51,49,51,52,50,49]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]

plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather Data')
plt.plot(days,max_t, label = 'Max')
plt.plot(days,min_t, label = 'Min')
plt.plot(days,avg_t, label = 'Avg')
plt.legend(loc = 'best')
```

```python
# Types of charts-> arranging in order->dividing a plotting place into 4
import matplotlib.pyplot as plt
%matplotlib inline
names = ['Group_a','Group_b','Group_c']
values = [10,50,100]

plt.figure(figsize=(10,10))
plt.subplot(224) #2 rows 2 columns at position- 4
plt.title('Bar Chart',loc = 'right') #location of bar chart title towards right
plt.bar(names,values)
plt.subplot(221)
plt.title('Scatter Chart')
plt.scatter(names,values)
plt.subplot(223)
plt.title('Line Chart')
plt.plot(names,values)
plt.subplot(222)
plt.title('Pie Chart')
plt.pie(values,labels = names)
plt.suptitle('Catergorical Plotting')
plt.show()
```

```python
li = []
n = int(input(''))
for i in range(n):
    a = input().split()
    if a[0] == "insert":
        li.insert(a[1],a[2])
    elif a[0] == "print":
        print(li)
    elif a[0] == "remove":
        li.remove(a[1])
    elif a[0] == "append":
        li.append(a[1])
    elif a[0] == "sort":
        li.sort()
    elif a[0] == "pop":
        li.pop(a[1])
    elif a[0] == "reverse":
        li.sort(reverse = True)
```
