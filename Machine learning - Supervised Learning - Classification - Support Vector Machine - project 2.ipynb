# Classifying the digits corresponding to the target names.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Importing the load digits sklearn dataset
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
dir(digits)
```

```python
for i in range(10):
    plt.gray()
    plt.matshow(digits.images[i])
```

```python
digits.data
```

```python
digits.target
```

```python
len(digits.target)
```

```python
df = pd.DataFrame(digits.data,digits.target)
df.head()
```

```python
df['Target'] = digits.target
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Independent Variable
X = df.drop(['Target'], axis = 'columns')
X.head()
```

```python
# Dependent variable
Y = df.Target
Y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size = 0.3,random_state = 17)
```

```python
# Importing Support vector classifier
from sklearn.svm import SVC
```

```python
# Taking kernel or transformation as 'linear'
model = SVC(gamma = 'auto',kernel = 'linear',C = 1.0)
```

```python
# Gamma is defined as the sensitivity of best fit line or plane towards points that are far away
# Gamma value insensitive to far away points is High Gamma -> Lower accuracy but more accomodating towards data variation
# Gamma value sensitive to far away points is Low Gamma -> Higher accuracy but even little variation in data is impactful
```

```python
# C value determines the regularization of best fit line or plane
# High regularization implies model is tending towards overfitting
# Low regularization implies line or plane is more comfortable to towards data points and could be less accurate
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
# Taking kernel or transformation as 'rbf'
rbf_model = SVC(kernel = 'rbf', gamma = 'auto')
```

```python
rbf_model.fit(X_train,y_train)
```

```python
rbf_model.score(X_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix
```

```python
y_pred = model.predict(X_test)
```

```python
con = cm(y_test,y_pred)
con
```

```python
import seaborn as sn
plt.figure(figsize = (10,5))
sn.heatmap(con, annot = True)
```
