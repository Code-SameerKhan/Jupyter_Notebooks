```python
x = data.drop('bought_insurance',axis = 1)
y = data['bought_insurance']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size = 0.2)
```

```python
x_train_scaled = X_train.copy()
x_train_scaled['age'] = x_train_scaled['age']/100 # Scaling the age in between 0 - 1

x_test_scaled = X_test.copy()
x_test_scaled['age'] = x_test_scaled['age']/100
```

```python
# Binary Cross Entropy or Log Loss
# -1/n summation(1-n)y_true.log(y_predicted)+(1-y_true).log(1-y_predicted)
```

```python
def log_loss(y_true,y_predicted):
    epsilon = 1e-15
    y_predicted_new = [max(i,epsilon) for i in y_predicted]
    y_predicted_new = [min(i,1-epsilon) for i in y_predicted_new]
    y_predicted_new = np.array(y_predicted_new)
    return -np.mean(y_true*np.log(y_predicted_new)+(1-y_true)*np.log(1-y_predicted_new))
```

```python
def sigmoid_numpy(X):
    return 1/(1+np.exp(-X))
```

```python
def gradient_descent(age,affordibility,y_true,epochs):
    # w1,w2,bias
    w1 = w2 = 1 # initialising values of weights as 1
    bias = 0 # initialising bias as 0
    learning_rate = 0.05
    n = len(age) # number of items in a column
    
    for i in range(epochs):
        weighted_sum = w1*age + w2*affordibility + bias
        y_predicted = sigmoid_numpy(weighted_sum)
        
        loss = log_loss(y_true,y_predicted)
        
        w1d =  (1/n)*np.dot(np.transpose(age),(y_predicted-y_true)) # Derivative of w1
        w2d = (1/n)*np.dot(np.transpose(affordibility),(y_predicted-y_true)) # Derivative of w2
        bd = np.mean(y_predicted-y_true)
        
        w1 = w1 - learning_rate * w1d # updating value of w1
        w2 = w2 - learning_rate * w2d # updating value of w2
        bias = bias - learning_rate * bd # updating value of bias
        
        print(f'Epoch:{i}, w1:{w1}, w2:{w2}, bias:{bias}, loss:{loss}')
        
    return w1, w2, bias
```

```python
gradient_descent(x_train_scaled['age'],x_train_scaled['affordibility'],y_train,1000)
```

```python
model = keras.Sequential([
    keras.layers.Dense(1,input_shape=(2,),activation ='sigmoid',kernel_initializer='ones',bias_initializer='zeros')
])
model.compile(optimizer = 'adam',
             loss = 'binary_crossentropy',
             metrics = ['accuracy'])
model.fit(x_train_scaled,y_train,epochs = 5000)
```

```python
model.evaluate(x_test_scaled,y_test)
```
